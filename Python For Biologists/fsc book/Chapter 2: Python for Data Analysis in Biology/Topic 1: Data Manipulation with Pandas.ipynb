```python
import pandas as pd
import numpy as np

# Create a DataFrame with 10 kinds of bacteria, each having at least 10 samples
data = {
    'Bacteria Type': ['E. coli', 'S. aureus', 'L. acidophilus', 'B. subtilis', 'P. aeruginosa',
                      'M. tuberculosis', 'C. difficile', 'H. pylori', 'S. typhimurium', 'K. pneumoniae'] * 10,
    'Sample ID': [f'Sample_{i+1}' for i in range(100)],
    'Growth Rate': [round(np.random.uniform(0.5, 2.5), 2) for _ in range(100)],  # Random growth rates between 0.5 and 2.5
    'Resistance Level': [np.random.choice(['Low', 'Medium', 'High']) for _ in range(100)]  # Random resistance levels
}

df_bacteria = pd.DataFrame(data)

# Shuffle the DataFrame to mix up the rows
df_bacteria = df_bacteria.sample(frac=1).reset_index(drop=True)

df_bacteria#.head(20)  # Display the first 20 rows to check
```

```python
# how many unique bacteria type s we have sing df_bacteria['DT].unique()
```

```python
df_bacteria['Bacteria Type'].unique()
```

```python
# find average growth rate for each bacteria type
```

```python
df_bacteria.groupby(by=['Bacteria Type']).mean().plot(kind='bar')
```

```python
# now find the bacteria based on the max value of grown 
```

```python
df_bacteria.groupby(by=['Bacteria Type']).max()#.plot(kind='bar')
```

```python
import plotly.graph_objects as go

# Assuming you have already loaded the data into 'female' and 'male' DataFrames

# Create traces for both genders
trace_female = go.Scatter(x=female.index, y=female['math score'], mode='lines', name='female')
trace_male = go.Scatter(x=male.index, y=male['math score'], mode='lines', name='male')

# Create the figure
fig = go.Figure()

# Add the traces to the figure
fig.add_trace(trace_female)
fig.add_trace(trace_male)

# Update the layout with axis labels and title
fig.update_layout(
    title='Math Scores by Gender',
    xaxis_title='Index',
    yaxis_title='Math Score',
    legend_title='Gender'
)

# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go

# Assuming you have 'female' and 'male' DataFrames with 'math score' columns
# Replace 'female' and 'male' with your actual DataFrames

# Create Scatter Trace for female math scores
trace_female = go.Scatter(
    x=list(range(len(female['math score']))),
    y=female['writing score'],
    mode='markers',
    name='Female'
)

# Create Scatter Trace for male math scores
trace_male = go.Scatter(
    x=list(range(len(male['math score']))),
    y=male['writing score'],
    mode='markers',
    name='Male'
)

# Create the layout for the plot
layout = go.Layout(
    title='Writing Scores by Gender',
    xaxis=dict(title='Student Index'),
    yaxis=dict(title='Math Score'),
)

# Combine the traces and layout and create the figure
fig = go.Figure(data=[trace_female, trace_male], layout=layout)

# Show the plot
fig.show()
```

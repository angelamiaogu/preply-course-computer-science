```python
def paralbol(x,a,h,k):
    return a*(x-h)**2 + k
```

```python

import numpy as np
x = np.arange(-100,100)
z = paralbol(x,1,0,0)
```

```python
import matplotlib.pyplot as plt
plt.plot(x,z)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the hyperbola
a = 5  # semi-major axis
b = 3  # semi-minor axis

# Range for x
x = np.linspace(-10, 10, 400)

# Calculate corresponding y values for the hyperbola
y_positive = np.sqrt((x**2 / a**2 - 1) * b**2)
y_negative = -np.sqrt((x**2 / a**2 - 1) * b**2)

# Plotting the hyperbola
plt.figure(figsize=(10, 6))
plt.plot(x, y_positive, 'b', label='y = sqrt((x^2 / a^2 - 1) * b^2)')
plt.plot(x, y_negative, 'r', label='y = -sqrt((x^2 / a^2 - 1) * b^2)')

# Adding asymptotes
asymptote_y1 = (b/a) * x
asymptote_y2 = -(b/a) * x
plt.plot(x, asymptote_y1, 'g--', label='y = (b/a) * x')
plt.plot(x, asymptote_y2, 'g--', label='y = -(b/a) * x')

# Setting labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Hyperbola: x^2/a^2 - y^2/b^2 = 1')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

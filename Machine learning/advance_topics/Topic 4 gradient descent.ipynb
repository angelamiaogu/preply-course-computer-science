```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample signal with noise
t = np.linspace(0, 1, 1000, endpoint=False)  # Time values
signal = 5 * np.sin(2 * np.pi * 10 * t) + 2 * np.random.randn(1000)  # Signal + noise

# Perform Fast Fourier Transform (FFT)
fft_result = np.fft.fft(signal)

# Set a threshold to remove high-frequency noise
threshold = 50
fft_result[threshold:-threshold] = 0

# Perform Inverse Fast Fourier Transform (IFFT)
filtered_signal = np.fft.ifft(fft_result)

# Plot the original signal and the filtered signal
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(t, signal, label='Original Signal')
plt.title('Original Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, filtered_signal, label='Filtered Signal', color='orange')
plt.title('Filtered Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample signal with noise
t = np.linspace(0, 1, 1000, endpoint=False)  # Time values
signal = 5 * np.sin(2 * np.pi * 10 * t) + 2 * np.random.randn(1000)  # Signal + noise

# Perform Fast Fourier Transform (FFT)
fft_result = np.fft.fft(signal)

# Set a threshold to remove high-frequency noise
threshold = 50
fft_thresholded = fft_result.copy()
fft_thresholded[threshold:-threshold] = 0

# Perform Inverse Fast Fourier Transform (IFFT) on thresholded FFT
filtered_signal = np.fft.ifft(fft_thresholded)

# Plotting multiple figures to illustrate the process
plt.figure(figsize=(12, 10))

# Plot 1: Original Signal
plt.subplot(4, 1, 1)
plt.plot(t, signal, label='Original Signal', color='blue')
plt.title('Original Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

# Plot 2: FFT Result
plt.subplot(4, 1, 2)
plt.plot(np.abs(fft_result), label='FFT Result', color='green')
plt.title('FFT Result')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.legend()

# Plot 3: Thresholded FFT
plt.subplot(4, 1, 3)
plt.plot(np.abs(fft_thresholded), label='Thresholded FFT', color='red')
plt.title('Thresholded FFT')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.legend()

# Plot 4: Filtered Signal
plt.subplot(4, 1, 4)
plt.plot(t, filtered_signal, label='Filtered Signal', color='orange')
plt.title('Filtered Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

def objective_function(x):
    return x**2 + 5*x + 10  # Quadratic function as an example

def gradient(x):
    return 2*x + 5  # Gradient of the quadratic function

def gradient_descent(initial_x, learning_rate, iterations):
    x_values = [initial_x]
    
    for _ in range(iterations):
        current_x = x_values[-1]
        gradient_value = gradient(current_x)
        new_x = current_x - learning_rate * gradient_value
        x_values.append(new_x)

    return x_values

# Control parameters
initial_x = 0.0
learning_rate = 0.1
iterations = 20

# Run gradient descent
x_values = gradient_descent(initial_x, learning_rate, iterations)

# Plotting
x_values_np = np.array(x_values)
y_values = objective_function(x_values_np)

plt.plot(x_values_np, y_values, marker='o')
plt.title('Gradient Descent: Learning Rate = {}'.format(learning_rate))
plt.xlabel('X')
plt.ylabel('Objective Function Value')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def objective_function(x):
    return x**2 + 5*x + 10  # Quadratic function as an example

def gradient(x):
    return 2*x + 5  # Gradient of the quadratic function

def gradient_descent(initial_x, learning_rate, iterations):
    x_values = [initial_x]
    
    for _ in range(iterations):
        current_x = x_values[-1]
        gradient_value = gradient(current_x)
        new_x = current_x - learning_rate * gradient_value
        x_values.append(new_x)

    return x_values

# Control parameters
initial_x = 0.0
learning_rate = 0.005
iterations = 100

# Run gradient descent
x_values = gradient_descent(initial_x, learning_rate, iterations)

# Plotting with different colors
x_values_np = np.array(x_values)
y_values = objective_function(x_values_np)

plt.plot(x_values_np, y_values, marker='o', color='b', label='Iterations')

# Highlight initial and final points
plt.scatter(x_values_np[0], y_values[0], color='g', marker='o', label='Initial Point')
plt.scatter(x_values_np[-1], y_values[-1], color='r', marker='o', label='Final Point')

plt.title('Gradient Descent: Learning Rate = {}'.format(learning_rate))
plt.xlabel('X')
plt.ylabel('Objective Function Value')
plt.legend()
plt.show()
```

```python
### 
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Define the objective function
def objective_function(x):
    return x**2 + 5*x + 10  # Quadratic function as an example

# Run optimization to find the local minimum
result = minimize(objective_function, x0=0.0, method='BFGS')

# Extract the optimal solution
optimal_x = result.x[0]
optimal_y = result.fun

# Plotting
x_values = np.linspace(-10, 5, 100)
y_values = objective_function(x_values)

plt.plot(x_values, y_values, label='Objective Function')
plt.scatter(optimal_x, optimal_y, color='r', marker='o', label='Local Minimum')

plt.title('Local Minimum of the Objective Function')
plt.xlabel('X')
plt.ylabel('Objective Function Value')
plt.legend()
plt.show()

print(f"Local minimum found at x = {optimal_x}, y = {optimal_y}")
```

```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Define a function with multiple local minima
def objective_function(x):
    return x**4 - 8*x**3 + 18*x**2

# Plotting the objective function
x_values = np.linspace(-2, 5, 100)
y_values = objective_function(x_values)

plt.plot(x_values, y_values, label='Objective Function')

# Run optimization with different initial guesses
initial_guesses = [-1, 0, 2, 4]

for guess in initial_guesses:
    result = minimize(objective_function, x0=guess, method='BFGS')
    optimal_x = result.x[0]
    optimal_y = result.fun
    plt.scatter(optimal_x, optimal_y, marker='x', label=f'Local Minima - Initial Guess: {guess:.1f}')

plt.title('Multiple Local Minima of the Objective Function')
plt.xlabel('X')
plt.ylabel('Objective Function Value')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Define a function with multiple local minima
def objective_function(x):
    return x**4 - 8*x**3 + 18*x**2

# Plotting the objective function
x_values = np.linspace(-2, 5, 100)
y_values = objective_function(x_values)

plt.plot(x_values, y_values, label='Objective Function')

# Run optimization with different initial guesses
initial_guesses = [-1, 0, 2, 4]
local_minima = []

for guess in initial_guesses:
    result = minimize(objective_function, x0=guess, method='BFGS')
    optimal_x = result.x[0]
    optimal_y = result.fun
    local_minima.append((optimal_x, optimal_y))
    plt.scatter(optimal_x, optimal_y, marker='o', label=f'Local Minimum - Initial Guess: {guess:.1f}')

plt.title('Multiple Local Minima of the Objective Function')
plt.xlabel('X')
plt.ylabel('Objective Function Value')
plt.legend()
plt.show()

print("Local minima found:")
for minimum in local_minima:
    print(f"At x = {minimum[0]:.4f}, y = {minimum[1]:.4f}")
```

```python
def distance(d,c):
    return np.sqrt (  (d[0] - c[0])**2 + (d[1]-c[1])**2  )

cent1 = [5,6]
cent2 = [3,4]
```

```python
for each in data:
    print(f'for point {each} c1 distance is {distance(each,cent1)} \t\t for point {each} c2 distance is {distance(each,cent2)}')
```

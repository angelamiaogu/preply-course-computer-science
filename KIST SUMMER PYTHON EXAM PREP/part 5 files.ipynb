```python
def banana(symbol, num):
    s = ''
    for i in range(num):
        for j in range(i):
            print(symbol,end='')
            s = s+symbol
        print()
        s = s+'\n'
    for a in range(num):
        for b in range(num-a-1,0,-1):
            print(symbol, end="")   
            s = s+symbol 
        print()
        s = s+'\n'
    return s
```

```python
ang = banana("angela ",200)
```

```python
file = open("banana.txt", 'w')
file.write(ang)
```

```python
import numpy as np
```

```python
def angela(x):
    y = x.split(" (")
    return(y[0])

def angela2(b):
    a = b.split(" (")
    if len(a) ==2:
        s2 = a[1].split(")")[0]
        return s2
    else:
        return np.nan
```

```python
data['movie_name']=data['title'].apply(angela)
```

```python
data['movie_year']=data['title'].apply(angela2)
```

```python
gen_data = data['genres'].str.get_dummies(sep='|')
gen_data
```

```python
data2 = pd.concat([data,gen_data],axis=1)
data2
```
